```python
def days_diff(dfs,start_day,end_day):
    day_range=pd.date_range(start=start_day, end=end_day)
    diff={}
    dfs_clean=dfs
    busta={}
    for i in range(len(dfs)):
        busta[i]={}
        diff[i]=day_range.difference(dfs[i]['Date'])
        diff[i]=(pd.DatetimeIndex.to_series(diff[i]).reset_index()).drop('index',axis=1)
        diff[i].columns=['Date']
        for j in range(len(dfs)):
            dfs_clean[j]=dfs_clean[j][~dfs_clean[j].Date.isin(diff[i]['Date'])]
    for i in range(len(dfs)):
        dfs_clean[i]=dfs_clean[i].reset_index()
        dfs_clean[i]=dfs_clean[i].drop('index',axis=1)
    return dfs_clean
```

```python
def Yield_col(dfs):
    zeros=np.zeros(len(dfs[0]))
    Yield={}
    Close_arr={}
    for i in range(len(dfs)):
        Yield[i]=zeros
        
        Close_arr[i]=dfs[i]['Close'].to_numpy()
        Yieldie=np.zeros(len(Close_arr[i]))
        for j in range(len(Yieldie)-1):
            Yieldie[j+1]=(Close_arr[i][j+1]-Close_arr[i][j])/Close_arr[i][j]
        Yield[i]=Yieldie
        Yield[i][0]=None
    return Yield
```

```python
def Arith_mean(rs):
    mean={}
    for i in range(len(rs)):
        mean[i]=np.sum(rs[i][1:])/len(rs[i][1:])
    return mean
```

```python
def Arith_mean_annualised(rs):
    mean={}
    for i in range(len(rs)):
        mean[i]=np.sum(rs[i][1:])/len(rs[i][1:])
        mean[i]=(1+mean[i])**252-1
    return mean
```

```python
def Geo_mean(rs):
    mean={}
    rsone={}
    for i in range(len(rs)):
        rsone[i]=rs[i]+1
        mean[i]=np.power(np.absolute(np.prod(rsone[i][1:],axis=0)),(1/(len(rsone[i][1:]))))
        mean[i]=mean[i]-1
    return mean
```

```python
n=10000
n_ass=len(Geo_mean_annualised(rs))
mean_variance_pairs=[]
mean_STD_pairs=[]
weights_list=[]
tickers_list=[]
np.random.seed(42)
for i in range(n):
    weights=np.random.rand(n_ass)
#    weights[2]=0
#    weights[1]=0
    weights=weights/np.sum(weights)
    
    assets = np.random.choice(list(rdf.columns), n_ass, replace=False)
    
    portfolio_E_Variance = 0
    portfolio_E_Return = 0
    for i in range(n_ass-1):
        portfolio_E_Return += weights[i] * Mus[i]
        for j in range(n_ass-1):
            portfolio_E_Variance += weights[i] * weights[j] * corrs[i][j] * stds[i] * stds[j] * 252
            STD=np.sqrt(portfolio_E_Variance)
    mean_variance_pairs.append([portfolio_E_Return, portfolio_E_Variance])
    mean_STD_pairs.append([portfolio_E_Return, STD])
    weights_list.append(weights)
    tickers_list.append(assets)
```

```python
#-- Plot the risk vs. return of randomly generated portfolios
#-- Convert the list from before into an array for easy plotting
mean_variance_pairs = np.array(mean_variance_pairs)

risk_free_rate=0 #-- Include risk free rate here

fig = go.Figure()
fig.add_trace(go.Scatter(x=np.sqrt(mean_variance_pairs[:,1]), y=mean_variance_pairs[:,0], 
                      marker=dict(color=(mean_variance_pairs[:,0]-risk_free_rate)/(mean_variance_pairs[:,1]**0.5), 
                                  showscale=True, 
                                  size=7,
                                  line=dict(width=1),
                                  colorscale="RdBu",
                                  colorbar=dict(title="Sharpe Ratio")
                                 ), 
                      mode='markers',
                        text=[str(np.array(tickers_list[i])) + "" + str(np.array(weights_list[i]).round(2)) for i in range(len(tickers_list))]))
fig.update_layout(template='plotly_white',
                  xaxis=dict(title='Annualised Risk (Volatility)'),
                  yaxis=dict(title='Annualised Return'),
                  title='Sample of Random Portfolios',
                  width=850,
                  height=500)
fig.update_xaxes(range=[0, 0.4])
fig.update_yaxes(range=[0,0.2])
fig.update_layout(coloraxis_colorbar=dict(title="Sharpe Ratio"))
```

```python
#-- Create random portfolio weights and indexes
#- How many assests in the portfolio
n_assets = n_ass

mean_variance_pairs = []
weights_list=[]
tickers_list=[]

for i in range(10000):
    next_i = False
    while True:
        #- Choose assets randomly without replacement
        assets = np.random.choice(list(rdf.columns), n_assets, replace=False)
        #- Choose weights randomly ensuring they sum to one
        weights = np.random.rand(n_assets)
        weights = weights/sum(weights)

        #-- Loop over asset pairs and compute portfolio return and variance
        portfolio_E_Variance = 0
        portfolio_E_Return = 0
        for i in range(n_ass-1):
            portfolio_E_Return += weights[i] * Mus[i]
            for j in range(n_ass-1):
                portfolio_E_Variance += 252*weights[i] * weights[j] * corrs[i][j] * stds[i] * stds[j]

        #-- Skip over dominated portfolios
        for R,V in mean_variance_pairs:
            if (R > portfolio_E_Return) & (V < portfolio_E_Variance):
                next_i = True
                break
        if next_i:
            break

        #-- Add the mean/variance pairs to a list for plotting
        mean_variance_pairs.append([portfolio_E_Return, portfolio_E_Variance])
        weights_list.append(weights)
        tickers_list.append(assets)
        break
```
